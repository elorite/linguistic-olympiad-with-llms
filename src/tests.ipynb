```python
from evaluator import evaluate_predictions
from prompt_strategies import zero_shot
import json
import time
import re
from main import load_dataset, strip_prefix
```

```python
dataset = load_dataset()
```

```python
# find unique types of problems
problem_types = []
for problem in dataset:
    if problem['type'] not in problem_types:
        problem_types.append(problem['type'])

# print problem types
print(f"Problem types ({len(problem_types)}):")
for problem_type in problem_types:
    print(f"- {problem_type}")
```

```python
# find unique languages in the dataset
languages = []
for problem in dataset:
    if problem['name'] not in languages:
        languages.append(problem['name'])

# remove final numbers from languages
languages = [re.sub(r'\d+$', '', language) for language in languages]
unique_languages = set(languages)

# print languages
print(f"\nLanguages in the dataset ({len(unique_languages)}):")
for language in unique_languages:
    print(f"- {language}")
```

```python
# select all NOUN-ADJ problems
counter = 0
for problem in dataset:
    if problem['type'] == 'NOUN-ADJ':
        counter += 1
print(counter)
```
